```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import requests 
import yfinance as yf
```

```python
# https://codingandfun.com/retrieving-currency-pairs-from-yahoo-finance-with-python/ for research 
```

```python
start = "2021-01-02"
end = "2021-01-10"
# setting start and end dates
# start takes T-1 
```

```python
list_of_pairs = [
"EURUSD=X"
,"JPY=X"
,"GBPUSD=X"
,"AUDUSD=X"
,"NZDUSD=X"
,"CNY=X"
,"HKD=X"
,"SGD=X"
,"INR=X"
,"MXN=X"
,"PHP=X"
,"IDR=X"
,"MYR=X"
,"THB=X"
,"ZAR=X"
,"RUB=X"
]
#Creating a list of pairs 
#Pairs are taken from https://finance.yahoo.com/currencies/
```

```python
list_of_values = []
for i in list_of_pairs:
    deneme = yf.Ticker(i)
    hist = deneme.history(start=start,end=end)
    hist["Name"] = i
    list_of_values.append(hist)
list_of_values
```

```python
# Making a for loop for appending all different currency dataframes together 
for i in range(0,len(list_of_values)):
    list_of_values[0] = list_of_values[0].append(list_of_values[i])
list_of_values[0]
```

```python
result = list_of_values[0]
result = result[["Name","Close"]] # Taking only columns which has name and close values
result = result.rename(columns={"Close":"Value"}) # Renaming close column to Value
result.to_csv("result.csv")
result
```
